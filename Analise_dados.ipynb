```python
import MetaTrader5 as mt5
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ações que utilizaremos para analisar: 
# ABEV3 AZUL4 BBAS3 CIEL3 CPFE3 PETR4 

#periodo de tempo de analise = 30 dias com kandle 
```

```python
mt5.initialize()
```

```python
cod_acoes = ['ABEV3']
```

```python
data = pd.DataFrame()
```

```python

for i in cod_acoes: 
    result_dataFrame_acoes= mt5.copy_rates_from_pos(i, mt5.TIMEFRAME_M5, 0, 1000)
    data['close'] = result_dataFrame_acoes['close']
    data['tick_volume'] = result_dataFrame_acoes['tick_volume']
    
data
    
```

```python
plt.plot(data['close'])
```

```python
plt.plot(data['tick_volume'])
```

```python
X = data['close'].values.reshape(-1, 1)
Y = data['tick_volume'].values.reshape(-1, 1)
data_close_MinMax = MinMaxScaler().fit_transform(X)
data_tick_MinMax = MinMaxScaler().fit_transform(Y)

```

```python
mediaX = np.median(X)
mediaY = np.median(Y)
desvioX = np.std(X)
desvioY = np.std(Y)


 
plt.scatter(X,Y,c='black')
plt.scatter(mediaX,mediaY,c='blue')
print(desvioX)
print(desvioY)
```
